Function to create stiffness and mass matrices:

```python
%matplotlib inline
import matplotlib.pyplot as plt
from JSAnimation import IPython_display
from matplotlib import animation


import numpy as np

def create_matrices(n):
    "Create stiffness and mass matrices of size n x n"

    # Cell length
    h = 1.0/(float(n) - 1.0)
    
    # Create a zero matrix
    M = np.zeros((n, n))
    K = np.zeros((n, n))

    # Add components of small matrix k to K
    k = (1.0/h)*np.array([[1.0, -1.0], [-1.0, 1.0]])
    for i in range(len(K) - 1):
        K[i:i+2, i:i+2] += k

    m = h*np.array([[1.0/3.0, 1.0/6.0], [1.0/6.0, 1.0/3.0]])
    for i in range(len(M) - 1):
        M[i:i+2, i:i+2] += m        
        
    return M, K
```

Explore how largest eigenvale changes with mesh refinement

```python
from scipy.linalg import eigvalsh

max_evals = []
for n in [2, 4, 8, 16, 32]:
    # Create matrices
    M, K = create_matrices(n + 1) 

    # Get largest eigenvalue
    evals = eigvalsh(K, M)
    max_evals.append(np.max(evals))

    h = 1.0/(float(n))
    ratio = "-"
    if len(max_evals) > 1:
        ratio = max_evals[-1]/max_evals[-2]
    print("h, largest eigenvalue, relative increase: {}, {}, {}".format(h, max_evals[-1], ratio))
```

Create matrices and apply boundary conditions

```python
# Problem size
N = 20

# Create matrices
M, K = create_matrices(N) 

# Apply boudary conditions
M[0, :], K[0, :] = 0.0, 0.0
M[:, 0], K[:, 0] = 0.0, 0.0
M[0, 0], K[0, 0] = 1.0, 1.0
```

```python
# Solve generalised eigenvalue problem
evals = eigvalsh(K, M)

# Get largest eigenvalue
e_max = np.max(evals)
print("Largest eigenvalue: {} ".format(e_max))
```

### Unsteady heat equation

Solving

$$
M \dot{a} + K a  = b
$$

```python
def initial_condition(n):
    x = np.linspace(0.0, 1.0, n)
    return x, 0.5*np.tanh(20.0*(x - 0.5)) + 0.5


x, aa0 = initial_condition(50)
plt.xlabel('$x$')
plt.ylabel('$u_{0}$')
plt.title('Initial condition')
plt.plot(x, aa0);
    
```

```python
def animate(method):
    fig = plt.figure()
    ax = plt.axes(xlim=(0, 1), ylim=(0, 1))
    plt.xlabel('$x$')
    plt.ylabel('$u$')
    line, = ax.plot([], [], lw=2)

    def init():
        line.set_data(x, a0)
        return line,

    def animate_method(i):
        a = method(a0)    
        line.set_data(x, a)
        return line

    return animation.FuncAnimation(fig, animate_method, init_func=init, frames=200, interval=100, repeat=False)
```

```python
def backward_euler(a0):
    b_hat = M.dot(a0)
    b_hat[0] = 0.0
    a0[:] = np.linalg.solve(M +dt*K, b_hat)
    return a0

# Set intial condition
x, a0 = initial_condition(N)

# Step time step
dt = 10.0*dt_crit
animate(backward_euler)
```
